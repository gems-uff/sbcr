## SBCR tunning

```python
df = pd.read_excel(SBCR_DATA, engine='openpyxl')
df
```

```python
df_attributes = pd.read_csv(DATASET_STATS)[['chunk_id','chunk_size', 'solution_size', 'chunk_size_char', 'solution_size_char', 'v1_size_char', 'v2_size_char']]
df_attributes = df_attributes[df_attributes['chunk_id'].isin(df['chunk_id'].unique())]
df_attributes
```

```python
df = df.merge(df_attributes, on='chunk_id')
```

```python
df['configuration'] = list(zip(df['maximum_neighbors'], df['max_iterations_without_improvement'], df['ils_timeout']))
df = df.drop_duplicates(subset=['chunk_id', 'configuration'])
df
```

```python
df.groupby('configuration').count().sort_values('chunk_id')
```

#### Rank the parameters by number of victories for each instance

```python
test = df.groupby('configuration').agg(
    solution_sim_lcs=('solution_sim_lcs','mean'),
    mean_time=('time_seconds','mean')
).reset_index()
test.columns = ['configuration', 'mean_solution_sim_lcs_overall', 'mean_time_overall']
test
```

```python
df['configuration'] = list(zip(df['maximum_neighbors'], df['max_iterations_without_improvement'], df['ils_timeout']))

# find the top 1 configuration for each chunk
max_scores = df.groupby('chunk_id')['solution_sim_lcs'].transform('max')
best_configs = df[df['solution_sim_lcs'] == max_scores]

# calculate the number of vicories and average similarity for chunks where the configuration is top 1
victory_count = best_configs.groupby('configuration').agg(
    victories=('configuration', 'count'),
    mean_solution_sim_lcs=('solution_sim_lcs', 'mean')  # avg similarity for top1 chunks
).reset_index()

# calculate average similarity overall for each configuration
mean_similarity_overall = df.groupby('configuration').agg(
    solution_sim_lcs=('solution_sim_lcs','mean'),
    mean_time=('time_seconds','mean')
).reset_index()
mean_similarity_overall.columns = ['configuration', 'mean_solution_sim_lcs_overall', 'mean_time_overall']

result = victory_count.merge(mean_similarity_overall, on='configuration', how='left')
result_sorted = result.sort_values(by='victories', ascending=False)
result_sorted[['configuration', 'victories', 'mean_solution_sim_lcs', 'mean_solution_sim_lcs_overall', 'mean_time_overall']]
```

#### Rank the parameters by the ascending value of the sum of ranks (for each instance)

```python
# Sort the configurations by chunk_id and solution_sim_lcs (score)
df['rank'] = df.groupby('chunk_id')['solution_sim_lcs'].rank(method='min', ascending=False)

# Sum the ranks for each configuration
config_ranking = df.groupby('configuration').agg(
    rank_sum=('rank', 'sum'),
    rank_avg=('rank', 'mean'),
    mean_similarity=('solution_sim_lcs', 'mean'),
    min_similarity=('solution_sim_lcs', 'min'),
    max_similarity=('solution_sim_lcs', 'max'),
    median_similarity=('solution_sim_lcs', 'median'),
    mean_time=('time_seconds', 'mean')
).reset_index()

# Classify the configurations based on the sum of ranks (smallest sum is better)
config_ranking['overall_rank'] = config_ranking['rank_sum'].rank(method='min')

#sort by overall ranking
config_ranking_sorted = config_ranking.sort_values(by='overall_rank')

config_ranking_sorted[['configuration', 'rank_sum', 'rank_avg', 'overall_rank', 'mean_similarity','min_similarity', 'max_similarity', 'median_similarity', 'mean_time']]
```

```python
result = config_ranking_sorted[['configuration', 'rank_avg', 'mean_similarity', 'mean_time']].merge(result_sorted[['victories', 'configuration']],on='configuration' ,  how='outer')
result.loc[result['victories'].isnull(),'victories'] = 0
# result.to_csv('result.csv')
result
```
